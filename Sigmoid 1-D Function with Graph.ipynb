```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Sigmoid1dFun(x,w,b):
    return 1/(1 + np.exp(-(w*x+b)))
```

```python
Sigmoid1dFun(1,0.2,0)
```

```python
Sigmoid1dFun(2,0.2,0)
```

```python
w = -1
b = 0
X = np.linspace(-10,10,100)
Y = Sigmoid1dFun(X,w,b)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
w = -1
b = 3
X = np.linspace(-10,10,100)
Y = Sigmoid1dFun(X,w,b)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
w = 1
b = 0
X = np.linspace(-10,10,100)
Y = Sigmoid1dFun(X,w,b)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
w = 1
b = -2.5
X = np.linspace(-10,10,100)
Y = Sigmoid1dFun(X,w,b)
```

```python
plt.plot(X,Y)
plt.show()
```

### if we change w => Slope changes
### if we change b => Midpoint changes
