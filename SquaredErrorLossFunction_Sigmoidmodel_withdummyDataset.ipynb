```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib.colors
```

```python
def Sigmoid(x,w,b):
    return 1/(1 + np.exp(-(w*x+b)))
```

```python
X = np.linspace(-10,10,100)

w = 0.5
b = 0.5

Y = Sigmoid(X,w,b)
```

```python
plt.plot(X,Y,'*')
plt.show()
```

```python
def CalculateLoss(X,Y,wval,bval):
    loss = 0
    for x,y in zip(X,Y):
        loss = loss + (y - Sigmoid(x,wval,bval))**2
    return loss
```

```python
W = np.linspace(0,2,100)
B = np.linspace(0,2,100)

WW, BB = np.meshgrid(W,B)

Loss = np.zeros(WW.shape)
```

```python
for i in range(WW.shape[0]):
    for j in range(WW.shape[1]):
        Loss[i, j] = CalculateLoss(X, Y, WW[i,j],BB[i,j])
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(WW,BB,Loss,cmap='viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('Loss')
ax.view_init(30,270)
```
