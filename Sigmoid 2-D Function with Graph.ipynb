```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib.colors
```

```python
def Sigmoid2dFun(x1, x2, w1, w2, b):
    return 1/(1 + np.exp(-(w1*x1 + w2*x2 + b)))
```

```python
Sigmoid2dFun(1, 0.5, 1, 0.5, 0.5)
```

```python
X1 = np.linspace(-10,10,100)
X2 = np.linspace(-10,10,100)

XX1, XX2 = np.meshgrid(X1,X2)
```

```python
w1 = 1.5
w2 = -0.5
b = 0
Y = Sigmoid2dFun(XX1,XX2,w1,w2,b)
```

```python
plt.contourf(XX1,XX2,Y,cmap = "viridis", alpha=0.75)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(XX1, XX2, Y, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.view_init(30,270)
```
